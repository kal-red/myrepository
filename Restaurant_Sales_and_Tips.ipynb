# Plotting Countplot

```python
sns.countplot(data=dr1,x='sex',hue='smoker')
```

# Plotting Jointplot 

```python
sns.jointplot(data=dr1,x='total_bill',y='tip',hue='sex')
```

# Plotting Distribution of Data

```python
sns.displot(data=dr1,x='total_bill')
plt.title('Distribution of Total_Bill')
plt.ylabel('Total Count')
```

```python
sns.displot(data=dr1,x='tip')
plt.title('Distribution of Tips')
plt.ylabel('Total Count')
```
